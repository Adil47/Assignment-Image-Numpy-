```python
import numpy as np
import os 
from PIL import Image

imageArray=[]

for dirName,subDirList,fileList in os.walk('./Images'):
    for file in fileList:
        img=Image.open(dirName+"/"+file)
        img=img.resize((200,200))
        imgArr=np.array(img)
        imageArray.append(imgArr)

imageArray
```
